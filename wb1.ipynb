```python
import numpy as np
import matplotlib.pyplot as plt

n = np.random.randint(2, 20)
masiv = np.random.rand(n)
print(masiv)

sr_znach = np.mean(masiv)
Median = np.median(masiv)

# Выводы
if abs(sr_znach - Median) < 0.015:
    print('Распределения примерно симметрично')
elif sr_znach - Median >= 0.015:
    print('несколько больших значений тянут среднее вверх')
else:
    print('несколько маленьких значений тянут среднее вниз')

x = np.arange(len(masiv))
plt.scatter(x, masiv)
plt.xlabel('Индекс')
plt.ylabel('Значения')
plt.title('Точечная диаграмма')
plt.axhline(Median, color='red', linestyle='--', label='Медиана')
plt.axhline(sr_znach, color='green', linestyle='--', label='Среднее')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    v = np.sqrt(1 + np.exp(np.sqrt(x)) + np.cos(x**2))
    n = abs(1 - np.sin(x) * np.sin(x) * np.sin(x))
    res = (v / n) + np.log(abs(2 * x))
    return res

x = np.random.uniform(1, 10, 10)
x = np.sort(x)
masiv = f(x)

srez_x = x[:5]
srez_masiv = masiv[:5]

plt.plot(x, masiv)
plt.scatter(srez_x, srez_masiv)

plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# 1. Загружаем данные за 2021 год
tickers = ["AAPL", "MSFT", "GOOG"]

data = yf.download(
    tickers,
    start="2021-01-01",
    end="2022-01-01",
    auto_adjust=True,
    progress=False
)

# 2. Берем цены закрытия (там много столбцов лишних, берём только цену закрытия)
close = data["Close"]

# 3. Строим график
plt.figure(figsize=(12, 6))

for ticker in tickers:
    plt.plot(close.index, close[ticker], label=ticker)  # (дата, цена, название графика)

plt.title("Стоимость акций Apple, Microsoft и Google за 2021 год")
plt.xlabel("Дата")
plt.ylabel("Цена закрытия ($)")
plt.legend()   # подпись графиков
plt.grid(True) # сетка

plt.tight_layout()  # для красоты
plt.show()

# 4. Дополнительно: считаем доходность за год
returns = (close.iloc[-1] / close.iloc[0] - 1) * 100

print("\nДоходность за 2021 год (%):")
print(returns.round(2))
```
